```python
name = "final_input_signs/S_"
for m in range(n):

    f_name = name + str(m) + ".png"
    new_name= "Extracted/"+files[m]
    print(files[m])
    im = cv2.imread(new_name)

# let's downscale the image using new  width and height
    down_width = 290
    down_height = 126
    down_points = (down_width, down_height)
    resized_down = cv2.resize(im, down_points, interpolation= cv2.INTER_LINEAR)
    crop = resized_down[20:120, 15:265] 
    kernel = np.ones((5,5),np.float32)/25
    dst = cv2.filter2D(crop,-1,kernel)
    median = cv2.medianBlur(crop,5)
    blur = cv2.GaussianBlur(crop,(3,3),0)
    cv2.imwrite(f_name, crop)
# let's upscale the image using new  width and height
#up_width = 600
#up_height = 400
#up_points = (up_width, up_height)
#resized_up = cv2.resize(img, up_points, interpolation= cv2.INTER_LINEAR)

# Display images
#cv2.imshow('Resized Down by defining height and width', resized_down)
#cv2.waitKey()
#cv2.imshow('Resized Up image by defining height and width', resized_up)
#cv2.waitKey()


#press any key to close the windows

#down_width = 250
#down_height = 100
#down_points = (down_width, down_height)
#resized = cv2.resize(crop, down_points, interpolation= cv2.INTER_LINEAR)

#cv2.imshow('cropped image', crop)
#cv2.waitKey()
#cv2.destroyAllWindows()
# Read the image using imread function
#image = cv2.imread('image.jpg')
#cv2.imshow('Original Image', image)

```
